#### Load data

```python
df2 = pd.read_csv("/Users/pari/Developer/NN_TF_NLTK/Data/emotion_recognizer_svm.csv")
```

```python
data = df2["Content"]
```

```python
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
```

```python
words = []
for w in data:
    y = w.split(" ")
    for x in y:
        words.append(x)
```

```python
cleanWords = []
for w in words:
    if w not in stopWords:
        if len(w) > 2:
            cleanWords.append(w)
```

```python
emotiontfidVct = CountVectorizer()
y = emotiontfidVct.fit_transform(cleanWords)
```

```python
df2 = pd.DataFrame(y.toarray(), columns=emotiontfidVct.get_feature_names())
```
